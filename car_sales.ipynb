```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.xlabel("Units_sold",fontsize=14)
plt.ylim(0,300)
plt.plot(reshaped_df.index,reshaped_df["Alto 800"])
plt.plot(reshaped_df.index,reshaped_df["Baleno"])
plt.plot(reshaped_df.index,reshaped_df["Urus"])
plt.plot(reshaped_df.index,reshaped_df["Fortuner"])
plt.plot(reshaped_df.index,reshaped_df["Echo"])
plt.plot(reshaped_df.index,reshaped_df["Creta"])
plt.plot(reshaped_df.index,reshaped_df["XUV 300"])
plt.plot(reshaped_df.index,reshaped_df["Bolero"])
plt.plot(reshaped_df.index,reshaped_df["Maserati"])
plt.plot(reshaped_df.index,reshaped_df["Harier"])
plt.plot(reshaped_df.index,reshaped_df["Mustang"])

```

```python

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.xlabel("Units_Sold",fontsize=14)
plt.ylim(0,350)
for column in reshaped_df.columns:
    plt.plot (reshaped_df.index,reshaped_df[column],label=reshaped_df[column],linewidth=3)
plt.legend(fontsize=16)
plt.show(block=False)
```

```python

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.xlabel("Units_Sold",fontsize=14)
plt.ylim(0,350)
for column in reshaped_df.columns:
    plt.plot (reshaped_df.index,reshaped_df[column],linewidth=3,label=column)
plt.legend(fontsize=16)
```

```python
rolling_df = reshaped_df.rolling(window=6).mean()
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date",fontsize=14)
plt.ylabel("U",fontsize=14)
plt.ylim(0,35000)
for column in reshaped_df.columns:
    plt.plot(reshaped_df.index,rolling_df[column],label=reshaped_df[column],linewidth=3)
plt.legend(fontsize=16)
plt.show(block=False)
```
